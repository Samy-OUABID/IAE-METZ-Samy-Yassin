```python
# Importation des bibliothèques
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Ticker du fonds d’investissement
ticker = "0P0001D3JX.F"

# Télécharger les données historiques depuis Yahoo Finance
try:
    df = yf.download(ticker, start="2020-01-01", end="2024-03-09", auto_adjust=False)
except Exception as e:
    print(f"Erreur lors du téléchargement des données : {e}")
    exit()

# Vérifier les colonnes disponibles
print("Colonnes disponibles :", df.columns)

# Utiliser 'Close' au lieu de 'Adj Close' si nécessaire
if 'Adj Close' in df.columns:
    df['Price'] = df['Adj Close']
elif 'Close' in df.columns:
    df['Price'] = df['Close']
else:
    print("Erreur : La colonne 'Close' ou 'Adj Close' est introuvable.")
    exit()

# Ajouter la colonne 'Return' (rendement journalier)
df['Return'] = df['Price'].pct_change()

# Supprimer les valeurs NaN générées par le calcul des rendements
df = df.dropna()

# Afficher les premières lignes après transformation
print(df.head())

# ✅ 1. VISUALISATION : Distribution des prix de clôture
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], kde=True)
plt.title('Distribution des Prix de Clôture du Fonds')
plt.xlabel('Prix de Clôture')
plt.ylabel('Fréquence')
plt.show()

# ✅ 2. MATRICE DE CORRÉLATION
plt.figure(figsize=(10, 6))
sns.heatmap(df[['Price', 'Return']].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matrice de Corrélation')
plt.show()

# ✅ 3. CALCUL DES INDICATEURS DE PERFORMANCE
risk_free_rate = 0.02  # Taux sans risque (2%)

# Calcul du rendement annualisé
annualized_return = df['Return'].mean() * 252  # 252 jours de trading par an

# Calcul de la volatilité annualisée (écart-type)
annualized_std = df['Return'].std() * np.sqrt(252)

# Calcul du ratio de Sharpe
sharpe_ratio = (annualized_return - risk_free_rate) / annualized_std
print(f'Ratio de Sharpe: {sharpe_ratio}')

# Calcul du ratio de Sortino (ne prend en compte que la volatilité négative)
downside_returns = df.loc[df['Return'] < 0, 'Return']
annualized_downside_std = downside_returns.std() * np.sqrt(252)
sortino_ratio = (annualized_return - risk_free_rate) / annualized_downside_std
print(f'Ratio de Sortino: {sortino_ratio}')

# ✅ 4. CLUSTERING DES FONDS D'INVESTISSEMENT

# Standardiser les données pour le clustering
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['Price', 'Return']])

# ACP : Réduction de dimensionnalité
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# Déterminer le nombre optimal de clusters avec le score de silhouette
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(pca_data)
    score = silhouette_score(pca_data, kmeans.labels_)
    silhouette_scores.append(score)

# Visualisation du score de silhouette
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Scores de Silhouette pour différents nombres de clusters')
plt.xlabel('Nombre de Clusters')
plt.ylabel('Silhouette Score')
plt.show()

# Appliquer K-Means avec le nombre optimal (ex: 3 clusters)
optimal_clusters = 3
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans.fit(pca_data)
df['Cluster'] = kmeans.labels_

# Visualisation des clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_data[:, 0], y=pca_data[:, 1], hue=df['Cluster'], palette='viridis')
plt.title('Clusters des Fonds d’Investissement')
plt.xlabel('Composante Principale 1')
plt.ylabel('Composante Principale 2')
plt.legend(title='Cluster')
plt.show()

# ✅ 5. RÉSUMÉ DES CLUSTERS
cluster_summary = df.groupby('Cluster').mean()
print(cluster_summary)
```
